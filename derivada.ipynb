```python
import math 
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
from importlib import reload
plt=reload(plt)
%matplotlib inline


def f(x):
    return ((x**3)*(np.sin(x)))

def df_analytic(x):
    return( 3*(x**2)*(np.sin(x)) +  (x**3)*(np.cos(x)) )

def df_direit(dx,x):
    return  ((f(x+dx) - f(x))/dx)

def df_cent(dx,x):
    return (f(x+dx) - f(x-dx))/(2*dx)

def error_df_direit(dx,x):
    return (abs(df_direit(dx,x) - df_analytic(x))/abs(df_analytic(x)))

def error_df_central(dx,x):
    return (abs(df_cent(dx,x) - df_analytic(x))/abs(df_analytic(x)))

x=3
dx = np.logspace(-16,0)
df_direita = []
df_centrada = []
df_analitica = []
erro_direita = []
erro_central = []
for i in dx:
    #adireita = (f(x+i) - f(x))/i
    #central = (f(x+i) - f(x-i))/(2*i)
    #df_direita.append(adireita)
    #df_centrada.append(central)
    
    df_direita.append(df_direit(i,x))
    df_centrada.append(df_cent(i,x))
    erro_direita.append(error_df_direit(i,x))
    erro_central.append(error_df_central(i,x))
    
    #df_direita.append(derivative(f,x,i))
    #df_centrada.append(derivative(f,x,i))
    df_analitica.append(df_analytic(x))
    

    
plt.plot(dx,df_direita,label="f' à direita")
plt.plot(dx,df_centrada,label="f' centrada")
plt.plot(dx,df_analitica,label="f' analitica")
plt.xlabel("Delta x")
plt.ylabel("f'(x)")
plt.xscale('log')
plt.legend()
plt.savefig("derivative.png")
```

```python
##erro
plt.plot(dx,erro_direita,label="erro à direita")
plt.plot(dx,erro_central,label="erro centrada")
plt.xscale('log')
plt.xlabel("Delta x")
plt.ylabel("erro")
plt.legend()
plt.savefig("error.png")
```
