```python
#Atividade projetil
v_o = 100/3.6
y_o=1
g = 9.81
x_o=0
def y_theta(x):
    x_o=50
    return (x_o - ((v_o*np.cos(x))/g)*( v_o*np.sin(x) + np.sqrt( (v_o**2)*((np.sin(x))**2) +  2*g*y_o) ) )
    
#a
x = np.arange(0,90,0.5)
plt.plot(np.radians(x),y_theta(np.radians(x)))
```

```python
root1 = newton(y_theta,0.3,1e-12)
root2 = newton(y_theta,1.2,1e-12)
print("raiz1 em graus: {0}".format(np.degrees(root1)))
print("raiz2 em graus: {0}".format(np.degrees(root2)))
```

```python
#graficos

def x(t,theta):
    return (x_o + v_o*np.cos(theta)*t)
def y(t,theta):
    return (y_o + v_o*np.sin(theta)*t - (1/2)*g*t**2)

t = np.arange(0,10,0.1)
plt.plot(x(t,root1),y(t,root1),label='theta={0}'.format(np.degrees(root1)))
plt.plot(x(t,root2),y(t,root2),label='theta={0}'.format(np.degrees(root2)))
plt.xlabel("x")
plt.ylim(0)
plt.xlim(0,60)
plt.ylabel("y")
plt.legend()
plt.savefig("trajetoria.pdf")
```
