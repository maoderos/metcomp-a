```python
import numpy as np
import random
from math import *

## QUEST√ÉO 1

def return_array(N):
    Y = np.zeros(N)
    for i in range(N):
        x1 = random.randint(1,6)
        x2 = random.randint(1,6)
        Y[i] = x1 + x2 
    return Y

def mean(x):
    arr = np.shape(x)
    N = arr[0]
    soma = 0 
    for i in range(N):
        soma += x[i]
    return (soma/N)
    
def std_mean(x,x_mean):
    arr = np.shape(x)
    N = arr[0]
    variance = 0
    for i in range(N):
        variance += (x[i]- x_mean)**2
    return sqrt(variance/N)
    
    
    
    
y = return_array(100)
print(np.shape(y))
y_mean = mean(y)
mean = np.mean(y)
print("mean with function: {0} | mean with np: {1}".format(y_mean,mean))
y_std = std_mean(y,y_mean)
stdmean = np.std(y)
print("std_mean with function: {0} | std_mean with np: {1}".format(y_std,stdmean))

```
