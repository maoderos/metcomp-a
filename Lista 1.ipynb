```python
#1
def is_prime(N):
    if (N < 0):
        print("n é natural")
        return
    i = 2
    if (N==1):
        print("não é primo")
    while i < N:
        if (N % i == 0):
            print("não é primo")
            break
        i +=1
    
    
print("Digite um número natural:")
n = input()
is_prime(float(n))
```

```python
#2
def divisors(N,n1,n2):
    i1 = n1
    divisors = []
    while (i1 <= n2):
        if(N % i1 == 0):
            divisors.append(i1)
        i1 +=1
    print("divisores de {0} entre {1} e {2} são: {3}".format(N,n1,n2,divisors))

divisors(6,2,5)
```

```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
from importlib import reload
plt=reload(plt)

#8
def gaussiana(x,mu,sigma):
    return (1/(np.sqrt(2*np.pi*sigma)))*np.exp(-((x-mu)**2)/(2*sigma**2))
#a
x = np.arange(-10,10,0.01)
y = gaussiana(x,0,1)
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y,label='sigma=1')
plt.legend()
plt.show()

#b
x = np.arange(-10,10,0.01)
y = gaussiana(x,0,2)
plt.plot(x,y,label='sigma=2')
x = np.arange(-10,10,0.01)
y = gaussiana(x,0,0.5)
plt.plot(x,y,label='sigma=0.2')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

#c
x = np.arange(-10,10,0.01)
y = gaussiana(x,0,1)
plt.plot(x,y,label='mu=0')
x = np.arange(-10,10,0.01)
y = gaussiana(x,1,1)
plt.plot(x,y,label='mu=1')
x = np.arange(-10,10,0.01)
y = gaussiana(x,2,1)
plt.plot(x,y,label='mu=2')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

```
