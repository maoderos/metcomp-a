```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def binomial(n,k):
    if (n<k):
        print("n deve ser maior ou igual a k")
    
    i=1
    num=1
    dem=1
    while(i<=k):
        num = num*(n - i + 1)
        dem = dem*i
        i +=1
    return num/dem

print(int(binomial(4,2)))

y = []
x = []
n = 100
for k in range(101):
    y.append(binomial(n,k)/(2**n))
    x.append(k)

plt.plot(x,y)


```
